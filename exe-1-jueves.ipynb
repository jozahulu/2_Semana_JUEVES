```python
def distancia(x1,x2,y1,y2):         #Distancia en el x,y
    d = np.sqrt((x2 - x1)**2 + (y2-y1)**2)
    return d
```

```python
def dist_x(x1,x2):   #Distancia solo para X 
    d_x = np.sqrt((x2 - x1)**2 )
    return d_x
```

```python
def dist_y(y1,y2):  #Distancia solo para y
    d_y = np.sqrt((y2-y1)**2)
    return d_y
```

```python
def punto(x,y,v):  #Distancia solo para y
    p= x*y+v
    return p
```
